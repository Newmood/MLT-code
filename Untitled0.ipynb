```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.decomposition import KernelPCA

# Create non-linearly separable dataset
X, y = make_moons(n_samples=100, noise=0.1, random_state=42)

# Apply Kernel PCA
kernel_pca = KernelPCA(kernel='rbf', gamma=15, n_components=2)
X_kernel_pca = kernel_pca.fit_transform(X)

# Plot the original data
plt.figure(figsize=(8, 6))
plt.scatter(X[y == 0, 0], X[y == 0, 1], label='Class 0', alpha=0.8)
plt.scatter(X[y == 1, 0], X[y == 1, 1], label='Class 1', alpha=0.8)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Original Data')
plt.legend()
plt.show()

# Plot data after Kernel PCA
plt.figure(figsize=(8, 6))
plt.scatter(X_kernel_pca[y == 0, 0], X_kernel_pca[y == 0, 1], label='Class 0', alpha=0.8)
plt.scatter(X_kernel_pca[y == 1, 0], X_kernel_pca[y == 1, 1], label='Class 1', alpha=0.8)
plt.xlabel('Kernel Component 1')
plt.ylabel('Kernel Component 2')
plt.title('Data after Kernel PCA')
plt.legend()
plt.show()
```

```python
from sklearn.datasets import make_circles
from sklearn.decomposition import PCA, KernelPCA

# Create non-linearly separable data in 2D
X, y = make_circles(n_samples=200, noise=0.05, random_state=42)

# Plot the original data
plt.figure(figsize=(6, 6))
plt.scatter(X[y == 0, 0], X[y == 0, 1], label='Class 0', alpha=0.8)
plt.scatter(X[y == 1, 0], X[y == 1, 1], label='Class 1', alpha=0.8)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Original Data')
plt.legend()
plt.show()

# Apply 3rd-degree Kernel PCA to project into 3D feature space
kernel_pca = KernelPCA(kernel='poly', degree=3, n_components=3)
X_kernel_pca = kernel_pca.fit_transform(X)

# Plot data after 3rd-degree Kernel PCA
plt.figure(figsize=(8, 6))
plt.scatter(X_kernel_pca[y == 0, 0], X_kernel_pca[y == 0, 1], label='Class 0', alpha=0.8)
plt.scatter(X_kernel_pca[y == 1, 0], X_kernel_pca[y == 1, 1], label='Class 1', alpha=0.8)
plt.xlabel('Kernel Component 1')
plt.ylabel('Kernel Component 2')
plt.title('Data after 3rd-degree Kernel PCA')
plt.legend()
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

# Load Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Apply PCA for dimensionality reduction (from 4D to 2D)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Plot the data after PCA
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[y == 0, 0], X_pca[y == 0, 1], label='Setosa', alpha=0.8)
plt.scatter(X_pca[y == 1, 0], X_pca[y == 1, 1], label='Versicolor', alpha=0.8)
plt.scatter(X_pca[y == 2, 0], X_pca[y == 2, 1], label='Virginica', alpha=0.8)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Data after PCA (2D)')
plt.legend()
plt.show()
```

```python
from sklearn.datasets import make_swiss_roll
from sklearn.manifold import TSNE

# Create non-linear data in 3D (Swiss Roll)
X, y = make_swiss_roll(n_samples=1000, random_state=42)

# Apply t-SNE for visualization (from 3D to 2D)
tsne = TSNE(n_components=2)
X_tsne = tsne.fit_transform(X)

# Plot the data after t-SNE
plt.figure(figsize=(8, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap=plt.cm.get_cmap('rainbow', 10), alpha=0.8)
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.title('Data after t-SNE (2D)')
plt.colorbar()
plt.show()
```
